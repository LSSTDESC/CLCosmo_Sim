```python
import numpy as np
import os
import matplotlib.pyplot as plt
import scipy.integrate
from astropy.table import Table, join
import healpy
from clmm import Cosmology
import pyccl as ccl
import clmm
```

```python
#fiducial csmoDC2 values
Omega_m_true = 0.2648
Omega_b_true = 0.0448
Omega_c_true = Omega_m_true - Omega_b_true
sigma8_true = 0.8
H0_true = 71
ns_true = 0.963
True_value = [Omega_m_true, sigma8_true]
cosmo_clmm = Cosmology(H0 = H0_true, Omega_dm0 = Omega_c_true, Omega_b0 = Omega_b_true, Omega_k0 = 0.0)
cosmo_ccl = ccl.Cosmology(Omega_c = Omega_c_true, Omega_b = Omega_b_true, h = H0_true/100, sigma8 = sigma8_true, n_s=ns_true, w0=-1, wa=0)
cosmo_clmm.be_cosmo = cosmo_ccl
```

```python

z = np.linspace(0.015, 1, 10)
```

```python
Om_list = np.linspace(0.1, 0.6, 4)
cosmo_ccl = ccl.Cosmology(Omega_c = Omega_c_true, Omega_b = Omega_b_true, h = H0_true/100, sigma8 = sigma8_true, n_s=ns_true, w0=-1, wa=0)
cosmo_clmm.be_cosmo = cosmo_ccl
dz_fiz = cosmo_clmm.eval_da_z1z2(0.01, z)
for i, Om_ in enumerate(Om_list):
    cosmo_clmm.be_cosmo = ccl.Cosmology(Omega_c = Om_ - Omega_b_true, Omega_b = Omega_b_true, h = H0_true/100, sigma8 = sigma8_true, n_s=ns_true, w0=-1, wa=0)
    plt.plot(z, cosmo_clmm.eval_da_z1z2(0.01, z)/dz_fiz-1, '-' + 'C'+str(i))


w_list = np.linspace(-2.5, -0.5, 4)
for i, w_ in enumerate(w_list):
    cosmo_clmm.be_cosmo = ccl.Cosmology(Omega_c = Omega_c_true, Omega_b = Omega_b_true, h = H0_true/100, sigma8 = sigma8_true, n_s=ns_true, w0=w_, wa=0)
    plt.plot(z, cosmo_clmm.eval_da_z1z2(0.01, z)/dz_fiz-1, '--' + 'C'+str(i), lw=2)

str_Om = ''
for Om_ in Om_list:
    str_Om += f'{Om_:.2f}, '

str_w = ''
for w_ in w_list:
    str_w += f'{w_:.2f}, '
plt.plot([], [], '-k', label = 'varying Om: ' + str_Om)
plt.plot([], [], '--k', label = 'varying w: '+ str_w)
plt.ylabel('Distance(z) / DC2 fiducial distance(z) - 1')
plt.xlabel('z')
plt.legend()
```
