```python
import numpy as np
import corner
import getdist
from getdist import plots, MCSamples
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
#plt.switch_backend("Qt5Agg")
```

```python
t0 = np.load('/pbs/throng/lsst/users/cpayerne/CLCosmo_Sim/richness_mass_from_DC2/chains/analysis_N.pkl', allow_pickle = True)
names = ['1', '2', '3', '4', '5', '6']
labels = [r'\ln \lambda_0', r'\mu_z', r'\mu_m', r'\sigma_{\ln \lambda, 0}', r'\sigma_z', r'\sigma_m']
settings={'mult_bias_correction_order':0,'smooth_scale_2D':5, 'smooth_scale_1D':5, 'boundary_correction_order':0}
N = MCSamples(samples=t0[16500:],names = names, 
                labels = labels, 
                 settings=settings, 
                 label=f'Abundance')
t2 = np.load('/pbs/throng/lsst/users/cpayerne/CLCosmo_Sim/richness_mass_from_DC2/chains/analysis_WLxN.pkl', allow_pickle = True)

WLN = MCSamples(samples=t2[16500:],names = names, 
                labels = labels, 
                 settings=settings, 
                 label=r'$\Delta\Sigma_{\rm cl} \times$ Abundance')

t1 = np.load('/pbs/throng/lsst/users/cpayerne/CLCosmo_Sim/richness_mass_from_DC2/chains/analysis_WL.pkl', allow_pickle = True)

WL = MCSamples(samples=t1[16500:],names = names, 
                labels = labels, 
                 settings=settings, 
                 label=r'$\Delta\Sigma_{\rm cl}$')
```

```python
t0 = np.load('/pbs/throng/lsst/users/cpayerne/CLCosmo_Sim/richness_mass_from_DC2/chains/analysis_full_cosmology_N.pkl', allow_pickle = True)
names = ['1', '2', '3', '4', '5', '6','7','8']
labels = [r'\ln \lambda_0', r'\mu_z', r'\mu_m', r'\sigma_{\ln \lambda, 0}', r'\sigma_z', r'\sigma_m','\Omega_m', '\sigma_8']
settings={'mult_bias_correction_order':0,'smooth_scale_2D':3, 'smooth_scale_1D':5, 'boundary_correction_order':0}
Nfull = MCSamples(samples=t0[t0[:,3]>0][50000:],names = names, 
                labels = labels, 
                 settings=settings, 
                 label=f'Abundance')
t2 = np.load('/pbs/throng/lsst/users/cpayerne/CLCosmo_Sim/richness_mass_from_DC2/chains/analysis_full_cosmology_WLxN.pkl', allow_pickle = True)

WLNfull = MCSamples(samples=t2[t2[:,3]>0][50000:],names = names, 
                labels = labels, 
                 settings=settings, 
                 label=r'$\Delta\Sigma_{\rm cl} \times$ Abundance')

t1 = np.load('/pbs/throng/lsst/users/cpayerne/CLCosmo_Sim/richness_mass_from_DC2/chains/analysis_full_cosmology_WL.pkl', allow_pickle = True)

WLfull = MCSamples(samples=t1[t1[:,3]>0][50000:],names = names, 
                labels = labels, 
                 settings=settings, 
                 label=r'$\Delta\Sigma_{\rm cl}$')
```
