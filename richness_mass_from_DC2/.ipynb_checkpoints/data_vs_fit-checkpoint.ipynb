```python
import numpy as np
import corner
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.load('test.pkl', allow_pickle = True)
```

```python
plt.plot(t[:,0])
```

```python
fig, ax = plt.subplots(4,4, figsize=(10,10))
sigma1 = 1. - np.exp(-(1./1.)**2/2.)
sigma2 = 1. - np.exp(-(2./1.)**2/2.)
sigma3 = 1. - np.exp(-(3./1.)**2/2.)
labels = [r'$\ln \lambda_0$', r'$G_z$', r'$G_m$', r'$\sigma_{\ln \lambda}$']
fig = corner.corner(t[13000:],bins=30, fig=fig, levels = [sigma1, sigma2, ], 
                    label_kwargs={"fontsize": 20}, plot_datapoints = False,
                    labels = labels, smooth=True, smooth1d=True)
for ax in fig.get_axes():
    ax.tick_params(axis='both', labelsize=16)
```
