```python
%load_ext autoreload

%autoreload 2
import numpy as np
import corner, sys
import getdist
from getdist import plots, MCSamples
%matplotlib inline
import matplotlib.pyplot as plt
sys.path.append('../../modules/')
import numpy as np
#plt.switch_backend("Qt5Agg")
```

```python
names = ['1', '2', '3', '4', '5', '6']
labels = [r'\ln \lambda_0', r'\mu_z', r'\mu_m', r'\sigma_{\ln \lambda, 0}', r'\sigma_z', r'\sigma_m']
settings={'mult_bias_correction_order':0,'smooth_scale_2D':3, 'smooth_scale_1D':4, 'boundary_correction_order':0}
```

```python
n_cut=15000
name='1-halo=nfw+c-M={}_photoz=Truez_hmf=Despali16'
name1 = name.format('Prada12')
t = np.load(f'../chains/WLxN/MCMC_fit_m-r_WLxN_{name1}.pkl', allow_pickle = True)
WLxN_Prada12 = MCSamples(samples=t['flat_chains'][n_cut:],names = names, 
                    labels = labels, 
                     settings=settings, 
                     label='Prada12')
name1 = name.format('Diemer15')
t = np.load(f'../chains/WLxN/MCMC_fit_m-r_WLxN_{name1}.pkl', allow_pickle = True)
WLxN_Diemer15 = MCSamples(samples=t['flat_chains'][n_cut:],names = names, 
                    labels = labels, 
                     settings=settings, 
                     label='Diemer15')
name1 = name.format('Duffy08')
t = np.load(f'../chains/WLxN/MCMC_fit_m-r_WLxN_{name1}.pkl', allow_pickle = True)
WLxN_Duffy08= MCSamples(samples=t['flat_chains'][n_cut:],names = names, 
                    labels = labels, 
                     settings=settings, 
                     label='Duffy08')
name1 = name.format('Bhattacharya13')
t = np.load(f'../chains/WLxN/MCMC_fit_m-r_WLxN_{name1}.pkl', allow_pickle = True)
WLxN_Bhattacharya13 = MCSamples(samples=t['flat_chains'][n_cut:],names = names, 
                    labels = labels, 
                     settings=settings, 
                     label='Bhattacharya13')
```

```python
g = plots.get_subplot_plotter(subplot_size=1.)
g.settings.lab_fontsize = 20
g.settings.legend_fontsize = 15
range_ = {'1':[2.5,4], '2':[-2,2], '3':[1,4], '4':[0.1,1.2],'5':[-1.5,1.5], '6':[-0.5,0.5], }
g.triangle_plot([WLxN_Prada12, WLxN_Diemer15,WLxN_Duffy08,WLxN_Bhattacharya13], 
                filled=[False, False, False, False,False,], 
                contour_colors = ['b', 'g', 'r', 'k', 'k', 'k' ], 
                contour_ls=['--','-','-', '-', '-'],
                contour_lws=[1,1,1,1,1],shaded=False,
                legend_fontsize = 10,
                #param_limits=range_,
                fine_bins = 1,  marker_args = {'linewidth':10}, )
plt.savefig(f'../fig/mcmc_scaling_relation_impact_CM_relation.png', dpi = 300)
```

```python
name='1-halo=nfw+c-M=Diemer15_photoz={}_hmf=Despali16'
name1 = name.format('Truez')
t = np.load(f'../chains/WLxN/MCMC_fit_m-r_WLxN_{name1}.pkl', allow_pickle = True)
WLxN_truez = MCSamples(samples=t['flat_chains'][n_cut:],names = names, 
                    labels = labels, 
                     settings=settings, 
                     label='Sources with true redshifts')
name1 = name.format('flex')
t = np.load(f'../chains/WLxN/MCMC_fit_m-r_WLxN_{name1}.pkl', allow_pickle = True)
WLxN_flex = MCSamples(samples=t['flat_chains'][n_cut:],names = names, 
                    labels = labels, 
                     settings=settings, 
                     label='FlexZBoost')
name1 = name.format('BPZ')
t = np.load(f'../chains/WLxN/MCMC_fit_m-r_WLxN_{name1}.pkl', allow_pickle = True)
WLxN_BPZ= MCSamples(samples=t['flat_chains'][n_cut:],names = names, 
                    labels = labels, 
                     settings=settings, 
                     label='BPZ')
```

```python
g = plots.get_subplot_plotter(subplot_size=1.)
g.settings.lab_fontsize = 20
g.settings.legend_fontsize = 15
range_ = {'1':[2.5,4], '2':[-2,2], '3':[1,4], '4':[0.1,1.2],'5':[-1.5,1.5], '6':[-0.5,0.5], }
g.triangle_plot([WLxN_truez, WLxN_flex,WLxN_BPZ], 
                filled=[False, False, False, False,False,], 
                contour_colors = ['b', 'g', 'r', 'k', 'k', 'k' ], 
                contour_ls=['--','-','-', '-', '-'],
                contour_lws=[1,1,1,1,1],shaded=False,
                legend_fontsize = 10,
                #param_limits=range_,
                fine_bins = 1,  marker_args = {'linewidth':10}, )
plt.savefig(f'../fig/mcmc_scaling_relation_impact_PZ.png', dpi = 300)
```

```python
name='1-halo=nfw+c-M=Diemer15_photoz={}_hmf=Despali16'
name1 = name.format('Truez')
t = np.load(f'../chains/WLxN/MCMC_fit_m-r_WLxN_{name1}.pkl', allow_pickle = True)
WLxN_1h = MCSamples(samples=t['flat_chains'][n_cut:],names = names, 
                    labels = labels, 
                     settings=settings, 
                     label='one - halo term')
name1 = '1-halo=nfw+c-M=Diemer15_+2-halo_photoz={}_hmf=Despali16'.format('Truez')
t = np.load(f'../chains/WLxN/MCMC_fit_m-r_WLxN_{name1}.pkl', allow_pickle = True)
WLxN_2h = MCSamples(samples=t['flat_chains'][n_cut:],names = names, 
                    labels = labels, 
                     settings=settings, 
                     label='one + two halo terms')
```
