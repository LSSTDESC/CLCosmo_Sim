```python
t0 = np.load('/pbs/throng/lsst/users/cpayerne/CLCosmo_Sim/richness_mass_from_DC2/chains/manuscript_analysis_M_.pkl', allow_pickle = True)
M = MCSamples(samples=t0[17500:],names = names, 
                labels = labels, 
                 settings=settings, 
                 label=r'$\langle M \rangle$')
t2 = np.load('/pbs/throng/lsst/users/cpayerne/CLCosmo_Sim/richness_mass_from_DC2/chains/manuscript_analysis_N_.pkl', allow_pickle = True)

N = MCSamples(samples=t2[17500:],names = names, 
                labels = labels, 
                 settings=settings, 
                 label=r'$N$')
t1 = np.load('/pbs/throng/lsst/users/cpayerne/CLCosmo_Sim/richness_mass_from_DC2/chains/manuscript_analysis_WL_.pkl', allow_pickle = True)

WL = MCSamples(samples=t1[17500:],names = names, 
                labels = labels, 
                 settings=settings, 
                 label=r'$\langle \Delta\Sigma \rangle$')
print(len(t1))
t1 = np.load('/pbs/throng/lsst/users/cpayerne/CLCosmo_Sim/richness_mass_from_DC2/chains/manuscript_analysis_WLxN_.pkl', allow_pickle = True)
print(len(t1))
print(np.mean(t1[9000:], axis=0))
WLN = MCSamples(samples=t1[17500:],names = names, 
                labels = labels, 
                 settings=settings, 
                 label=r'$\langle \Delta\Sigma \rangle\ x\ N$')

t1 = np.load('/pbs/throng/lsst/users/cpayerne/CLCosmo_Sim/richness_mass_from_DC2/chains/manuscript_analysis_MxN_.pkl', allow_pickle = True)

MN = MCSamples(samples=t1[17500:],names = names, 
                labels = labels, 
                 settings=settings, 
                 label=r'$\langle M\rangle\ x\ N$')

fiducial = np.load('/pbs/throng/lsst/users/cpayerne/CLCosmo_Sim/richness_mass_from_DC2/plots/fiducial.pkl.npy')
fid = MCSamples(samples=fiducial,names = names, 
                labels = labels, 
                 settings=settings, 
                 label=r'$Fiducial$')
```
