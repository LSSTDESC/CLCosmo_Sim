```python
import numpy as np
import corner
import matplotlib.pyplot as plt
import numpy as np
import getdist
from getdist import plots, MCSamples
```

```python
t1 = np.load('test_1.pkl', allow_pickle = True)
names = ['1', '2', '3', '4']
labels = [r'\ln \lambda_0', r'G_z', r'G_m', r'\sigma_{\ln \lambda}']
settings={'mult_bias_correction_order':0,'smooth_scale_2D':5, 'smooth_scale_1D':6, 'boundary_correction_order':0}
sampleTinker08 = MCSamples(samples=t1[20000:],names = names, 
                labels = labels, 
                 settings=settings, 
                 label=f'$Tinker08$')

t2 = np.load('test_2.pkl', allow_pickle = True)
sampleDespali16 = MCSamples(samples=t2[20000:],names = names, 
                labels = labels, 
                 settings=settings, 
                 label=f'$Despali16$')
```

```python
plt.plot(t1[:,0])
plt.plot(t2[:,0])
```

```python
g = plots.get_subplot_plotter(subplot_size=1.)
g.triangle_plot([sampleTinker08, sampleDespali16], filled=True, contour_colors = ['r', 'b'], contour_ls='-',
                contour_lws=1.4,
                #param_limits = range_,
                       fine_bins = 1,  marker_args = {'linewidth':2}, )
plt.savefig('mcmc_mlambda.png', dpi = 300)
```
